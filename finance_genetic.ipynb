```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# The Big Ten
company_list = ["TSLA", "AAPL", "GOOG", "AMZN", "MSFT", "META", "NVDA", "INTC", "BRK-B", "UNH"]
company_names = ["TESLA", "APPLE", "GOOGLE", "AMAZON", "MICROSOFT", "META", "NVIDIA", "INTEL", "BERKSHIRE HATHAWAY", "UNITEDHEALTH GROUP"]

# Download stock data from yfinance
def download_stock(stock):
    df = yf.download(
        [stock],
        start="2010-01-01",
        end="2024-01-01",
        group_by="ticker",
        progress=False
    )[['Adj Close']]
    #df.round(4)
    df.insert(loc=0, column="Company", value=stock)
    return df

# Print the general information of the given stock
def info_stock(stock_name, stock_data):
    print(f'{stock_name}\'s Historical Data: \n\n {stock_data} \n\n')
    print(f'{stock_name}\'s Column Information: \n')
    print(stock_data.info(),'\n\n')
    print(f'{stock_name}\'s Statistic Summary: \n\n {stock_data.describe().round(4)}')
    print('-'*75)
    print()
```

```python
def update_with_predictions(stock_data, predictions):
    start_date = pd.Timestamp('2020-01-08')
    trading_days = stock_data.loc[start_date:].index

    print(f"Number of trading days: {len(trading_days)}, Number of predictions: {len(predictions)}")

    # Eğer işlem günü sayısı ve tahmin sayısı uyuşmuyorsa, bu bir sorun olabilir
    #if len(trading_days) != len(predictions):
    #    print("Warning: The number of trading days and predictions do not match!")

    # CSV dosyasındaki tahminleri ilgili tarihlerle eşleştir ve güncelle
    for i, predicted_value in enumerate(predictions):  # Burada düzeltme yapıldı
        if i < len(trading_days):
            # Güncelleme işlemini doğrulayın
            old_value = stock_data.at[trading_days[i], 'Adj Close']
            stock_data.at[trading_days[i], 'Adj Close'] = predicted_value
     #       print(f"Updating {trading_days[i]}: {old_value} to {predicted_value}")

    return stock_data
```

```python
stock_dict = {}
for stock in company_list:
    stock_data = download_stock(stock)  # Önceden tanımlı download_stock fonksiyonunu kullanarak veriyi indir
    predictions_csv = pd.read_csv(f'/content/drive/MyDrive/data_future_predict (1)/{stock}_Predictions.csv')  # Doğru dosya yolunu kullanın
    predictions = predictions_csv['Predicted'].values
    # Güncelleme işlemi için stock_data'nın bir kopyasını kullan
    stock_data_copy = stock_data.copy()
   # print(f"Original Data for {stock}:")
    #print(stock_data_copy.tail(3))  # Son üç kaydı yazdır
    updated_stock_data = update_with_predictions(stock_data_copy, predictions)
    # Güncelleme işleminden sonra DataFrame'i kontrol et
    #print(f"Updated Data for {stock}:")
    #print(updated_stock_data.tail(3))  # Son üç kaydı yazdır
    stock_dict[stock] = updated_stock_data  # Güncellenmiş kopyayı stock_dict sözlüğüne kaydet

#for stock_name, stock_data in stock_dict.items():
#    info_stock(stock_name, stock_data)
```

```python
import numpy as np
import pandas as pd
from tqdm import tqdm

def optimize_portfolio_normal_genetic(stock_dict, start_date, end_date, population_size=1000, num_generations=90, mutation_rate=0.10):
    """
    Hisse senetleri için portföy optimizasyonu yapar.

    Parametreler:
    - stock_dict: Hisse senetlerinin DataFrame'lerini içeren sözlük.
    - start_date: Analiz için başlangıç tarihi.
    - end_date: Analiz için bitiş tarihi.
    - population_size: Popülasyon büyüklüğü.
    - num_generations: Nesil sayısı.
    - mutation_rate: Mutasyon oranı.
    """

    combined_returns = pd.DataFrame()
    for stock, df in stock_dict.items():
        df_filtered = df.loc[start_date:end_date]
        combined_returns[stock] = df_filtered['Adj Close'].pct_change().dropna()

    num_assets = len(stock_dict)
    population = np.random.dirichlet(np.ones(num_assets), size=population_size)

    def fitness(weights, returns):
        port_return = np.sum(weights * returns.mean()) * 252
        port_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))
        sharpe_ratio = port_return / port_volatility
        return -sharpe_ratio

    for generation in tqdm(range(num_generations), desc="Evolving Generations"):
        fitness_scores = np.array([fitness(individual, combined_returns) for individual in population])
        best_indices = np.argsort(fitness_scores)[:population_size // 2]
        survivors = population[best_indices]

        new_population = []
        while len(new_population) < population_size:
            parents = np.random.choice(survivors.shape[0], 2, replace=False)
            cross_point = np.random.randint(1, num_assets)
            child = np.concatenate([survivors[parents[0], :cross_point], survivors[parents[1], cross_point:]])
            if np.random.rand() < mutation_rate:
                mutation_point = np.random.randint(num_assets)
                child[mutation_point] = np.random.rand()
                child /= np.sum(child)  # ağırlıkları normalize et
            new_population.append(child)
        population = np.array(new_population)

    best_index = np.argmin([fitness(individual, combined_returns) for individual in population])
    best_weights = population[best_index]
    port_volatility = np.sqrt(np.dot(best_weights.T, np.dot(combined_returns.cov() * 252, best_weights)))
    optimized_weights_dict = {stock: weight for stock, weight in zip(stock_dict.keys(), best_weights)}

    return optimized_weights_dict, port_volatility

# stock_dict örneği ve kullanımı
# stock_dict = {'AAPL': apple_df, 'MSFT': microsoft_df, ...}
# start_date = '2020-01-01'
# end_date = '2020-12-31'
# best_weights, port_volatility = optimize_portfolio(stock_dict, start_date, end_date)
```

```python
import numpy as np
import pandas as pd
from tqdm import tqdm

def optimize_portfolio_elitism_genetic(stock_dict, start_date, end_date, population_size=1000, num_generations=90, elite_size=50, mutation_rate=0.01):
    """
    Hisse senetleri için elitizm odaklı portföy optimizasyonu yapar.

    Parametreler:
    - stock_dict: Hisse senetlerinin DataFrame'lerini içeren sözlük.
    - start_date: Analiz için başlangıç tarihi.
    - end_date: Analiz için bitiş tarihi.
    - population_size: Popülasyon büyüklüğü.
    - num_generations: Nesil sayısı.
    - elite_size: Her nesilde doğrudan sonraki nesle aktarılacak elit birey sayısı.
    - mutation_rate: Mutasyon oranı. (Çok düşük tutulabilir veya tamamen kaldırılabilir)
    """

    combined_returns = pd.DataFrame()
    for stock, df in stock_dict.items():
        df_filtered = df.loc[start_date:end_date]
        combined_returns[stock] = df_filtered['Adj Close'].pct_change().dropna()

    num_assets = len(stock_dict)
    population = np.random.dirichlet(np.ones(num_assets), size=population_size)

    def fitness(weights, returns):
        port_return = np.sum(weights * returns.mean()) * 252
        port_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))
        sharpe_ratio = port_return / port_volatility
        return -sharpe_ratio

    for generation in tqdm(range(num_generations), desc="Evolving Generations"):
        fitness_scores = np.array([fitness(individual, combined_returns) for individual in population])
        best_indices = np.argsort(fitness_scores)[:elite_size]  # Elit bireylerin endekslerini bul
        elites = population[best_indices]  # Elit bireyleri seç
        survivors = population[np.argsort(fitness_scores)[:population_size // 2]]

        new_population = elites.tolist()  # Elit bireyleri yeni popülasyona doğrudan ekle
        while len(new_population) < population_size:
            parents = np.random.choice(survivors.shape[0], 2, replace=False)
            cross_point = np.random.randint(1, num_assets)
            child = np.concatenate([survivors[parents[0], :cross_point], survivors[parents[1], cross_point:]])
            if np.random.rand() < mutation_rate:
                mutation_point = np.random.randint(num_assets)
                child[mutation_point] = np.random.rand()
                child /= np.sum(child)  # ağırlıkları normalize et
            new_population.append(child)
        population = np.array(new_population)

    best_index = np.argmin([fitness(individual, combined_returns) for individual in population])
    best_weights = population[best_index]
    port_volatility = np.sqrt(np.dot(best_weights.T, np.dot(combined_returns.cov() * 252, best_weights)))
    optimized_weights_dict = {stock: weight for stock, weight in zip(stock_dict.keys(), best_weights)}


    return optimized_weights_dict, port_volatility
```

```python
import numpy as np
import pandas as pd
from tqdm import tqdm

def optimize_portfolio__more_mutation_genetic(stock_dict, start_date, end_date, population_size=1000, num_generations=90, mutation_rate=0.20):
    """
    Hisse senetleri için portföy optimizasyonunu genetik algoritma kullanarak gerçekleştirir.

    Parametreler:
    - stock_dict: Hisse senetlerinin DataFrame'lerini içeren sözlük.
    - start_date: Analiz için başlangıç tarihi.
    - end_date: Analiz için bitiş tarihi.
    - population_size: Popülasyon büyüklüğü.
    - num_generations: Nesil sayısı.
    - mutation_rate: Mutasyon oranı.
    """

    combined_returns = pd.DataFrame()
    for stock, df in stock_dict.items():
        df_filtered = df.loc[start_date:end_date]
        combined_returns[stock] = df_filtered['Adj Close'].pct_change().dropna()

    num_assets = len(stock_dict)
    population = np.random.dirichlet(np.ones(num_assets) * 0.5, size=population_size)

    def fitness(weights, returns):
        port_return = np.sum(weights * returns.mean()) * 252
        port_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights)))
        sharpe_ratio = port_return / port_volatility
        return -sharpe_ratio

    for generation in tqdm(range(num_generations), desc="Evolving Generations"):
        fitness_scores = np.array([fitness(individual, combined_returns) for individual in population])
        best_indices = np.argsort(fitness_scores)[:population_size // 2]
        survivors = population[best_indices]

        new_population = []
        while len(new_population) < population_size:
            parents = np.random.choice(survivors.shape[0], 2, replace=False)
            cross_points = np.sort(np.random.choice(np.arange(1, num_assets - 1), 2, replace=False))
            child = np.concatenate([survivors[parents[0], :cross_points[0]],
                                    survivors[parents[1], cross_points[0]:cross_points[1]],
                                    survivors[parents[0], cross_points[1]:]])
            if np.random.rand() < mutation_rate:
                mutation_point = np.random.randint(num_assets)
                child += np.random.normal(0, 0.1, num_assets)
                child[mutation_point] = np.random.rand()
                child = np.clip(child, 0, 1)
                child /= np.sum(child)
            new_population.append(child)
        population = np.array(new_population)

    best_index = np.argmin([fitness(individual, combined_returns) for individual in population])
    best_weights = population[best_index]
    port_volatility = np.sqrt(np.dot(best_weights.T, np.dot(combined_returns.cov() * 252, best_weights)))
    optimized_weights_dict = {stock: weight for stock, weight in zip(stock_dict.keys(), best_weights)}

    return optimized_weights_dict, port_volatility

# stock_dict örneği ve kullanımı:
# stock_dict = {'AAPL': apple_df, 'MSFT': microsoft_df, ...}
# start_date = '2020-01-01'
# end_date = '2020-12-31'
# best_weights, port_volatility = optimize_portfolio_with_genetic_algorithm(stock_dict, start_date, end_date)
```
