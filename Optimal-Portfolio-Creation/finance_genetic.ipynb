```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
# The Big Seven
company_list = ["TSLA", "AAPL", "GOOG", "AMZN", "MSFT", "META", "NVDA"]
company_names = ["TESLA", "APPLE", "MICROSOFT", "AMAZON", "MICROSOFT", "META", "NVIDIA"]

# Download stock data from yfinance
def download_stock(stock):
    df = yf.download(
        [stock],
        start="2010-01-01",
        end="2024-01-01",
        group_by="ticker",
        progress=False
    )
    df.round(4)
    df.insert(loc=0, column="Company", value=stock)
    return df

# Print the general information of the given stock
def info_stock(stock_name, stock_data):
    print(f'{stock_name}\'s Historical Data: \n\n {stock_data} \n\n')
    print(f'{stock_name}\'s Column Information: \n')
    print(stock_data.info(),'\n\n')
    print(f'{stock_name}\'s Statistic Summary: \n\n {stock_data.describe().round(4)}')
    print('-'*75)
    print()
```

```python
stock_dict = {}
for stock in company_list:
    stock_dict[stock] = download_stock(stock)

for stock_name, stock_data in stock_dict.items():
    info_stock(stock_name, stock_data)
```

```python
import numpy as np
import pandas as pd
  # tqdm'i import et
from tqdm import tqdm
# stock_dict sözlüğünüz varsayılan olarak mevcut ve her bir hisse senedi için DataFrame içeriyor.
# Örnek olarak, sadece bir hisse senedi için (NVDA) nasıl yapılacağını gösteriyorum.
# Gerçek senaryoda, tüm hisse senetleri için bu işlemi yapmalısınız.

# Tüm hisse senetleri için günlük getirileri birleştirin
# combined_returns DataFrame'inin her sütunu bir hisse senedinin günlük getirilerini içerecek şekilde olmalıdır.
combined_returns = pd.DataFrame()
for stock in stock_dict.keys():
    combined_returns[stock] = stock_dict[stock]['Close'].pct_change().dropna()

# Genetik algoritma parametreleri
population_size = 1000
num_generations = 90
mutation_rate = 0.10
num_assets = len(stock_dict)


population = np.random.dirichlet(np.ones(num_assets), size=population_size)


def fitness(weights, returns):
    # Portföy getirisini ve volatilitesini hesapla
    port_return = np.sum(weights * returns.mean()) * 252 #closed dan günlük kar 252 gün var yılda
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights))) # risk volatile göre
    # Sharpe oranını hesapla
    sharpe_ratio = port_return / port_volatility
    return -sharpe_ratio  # maksimize etmek için negatif değer


for generation in tqdm(range(num_generations), desc="Evolving Generations"):
    fitness_scores = np.array([fitness(individual, combined_returns) for individual in population])
    best_indices = np.argsort(fitness_scores)[:population_size // 2]
    survivors = population[best_indices]

    new_population = []
    while len(new_population) < population_size:
        parents = np.random.choice(survivors.shape[0], 2, replace=False)
        cross_point = np.random.randint(1, num_assets)
        child = np.concatenate([survivors[parents[0], :cross_point], survivors[parents[1], cross_point:]])
        if np.random.rand() < mutation_rate:
            mutation_point = np.random.randint(num_assets)
            child[mutation_point] = np.random.rand()
            child /= np.sum(child)  # ağırlıkları normalize et
        new_population.append(child)
    population = np.array(new_population)

best_index = np.argmin([fitness(individual, combined_returns) for individual in population])
best_weights = population[best_index]
port_volatility = np.sqrt(np.dot(best_weights.T, np.dot(combined_returns.cov() * 252, best_weights)))

print("Optimized Portfolio Volatility (Risk):", port_volatility)
print("Optimized Portfolio Weights:", best_weights)
```

```python
#daha çeşitlendirmeye çalışıyoruz

import numpy as np
import pandas as pd

# stock_dict sözlüğünüz varsayılan olarak mevcut ve her bir hisse senedi için DataFrame içeriyor.
# Örnek olarak, sadece bir hisse senedi için (NVDA) nasıl yapılacağını gösteriyorum.
# Gerçek senaryoda, tüm hisse senetleri için bu işlemi yapmalısınız.

# Tüm hisse senetleri için günlük getirileri birleştirin
# combined_returns DataFrame'inin her sütunu bir hisse senedinin günlük getirilerini içerecek şekilde olmalıdır.
combined_returns = pd.DataFrame()
for stock in stock_dict.keys():
    combined_returns[stock] = stock_dict[stock]['Close'].pct_change().dropna()

# Genetik algoritma parametreleri
population_size = 1000
num_generations = 90
mutation_rate = 0.20
num_assets = len(stock_dict)


population = np.random.dirichlet(np.ones(num_assets)* 0.5, size=population_size) # * 0.5 eklendi


def fitness(weights, returns):
    # Portföy getirisini ve volatilitesini hesapla
    port_return = np.sum(weights * returns.mean()) * 252 #closed dan günlük kar
    port_volatility = np.sqrt(np.dot(weights.T, np.dot(returns.cov() * 252, weights))) # risk volatile göre
    # Sharpe oranını hesapla
    sharpe_ratio = port_return / port_volatility
    return -sharpe_ratio  # maksimize etmek için negatif değer


for generation in tqdm(range(num_generations), desc="Evolving Generations"):
    fitness_scores = np.array([fitness(individual, combined_returns) for individual in population])
    best_indices = np.argsort(fitness_scores)[:population_size // 2]
    survivors = population[best_indices]

    new_population = []
    while len(new_population) < population_size:
        parents = np.random.choice(survivors.shape[0], 2, replace=False)
        cross_points = np.sort(np.random.choice(np.arange(1, num_assets - 1), 2, replace=False))
        child = np.concatenate([survivors[parents[0], :cross_points[0]],
                                survivors[parents[1], cross_points[0]:cross_points[1]], # more cross
                                survivors[parents[0], cross_points[1]:]])

        # Mutasyon mekanizmasını güçlendir
        if np.random.rand() < mutation_rate:
            mutation_point = np.random.randint(num_assets)
            child += np.random.normal(0, 0.1, num_assets)  # Normal dağılımlı mutasyon
            child[mutation_point] = np.random.rand()
            child = np.clip(child, 0, 1)  # Değerleri [0, 1] aralığında sınırla
            child /= np.sum(child)  # Ağırlıkları normalize et

        new_population.append(child)
    population = np.array(new_population)

best_index = np.argmin([fitness(individual, combined_returns) for individual in population])
best_weights = population[best_index]
port_volatility = np.sqrt(np.dot(best_weights.T, np.dot(combined_returns.cov() * 252, best_weights)))

print("Optimized Portfolio Volatility (Risk):", port_volatility)
print("Optimized Portfolio Weights:", best_weights)
```
